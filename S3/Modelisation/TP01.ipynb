```python
def ligne_polygonale(points, style="r-"):
    x = points[0, :]
    y = points[1, :]
    plt.plot(x, y, style)
    plt.show()
    
    return


points = np.random.rand(2, 2)
ligne_polygonale(points)
```

```python
A = np.random.rand(2, 1)
B = np.random.rand(2, 1)
I = np.array([
        (A[0] + B[0]) / 2,
        (A[1] + B[1]) / 2
])
```

```python
ligne_polygonale(I)
```
