# Ex 2

```python
n = 500
p = 0.4
X = [0, 1]
compte = np.array([0, 0])
```

```python
for k in range(n):
    U = np.random.rand()
    if U > p :
        compte[1] += 1
    else:
        compte[0] += 1
```

```python
freqX = compte / n
plt.title("Fréquence : {}".format(freqX))
plt.bar(X, freqX, width=0.2, color="red")
plt.show()
```

# Ex 3

```python
def plot_observation(n, X):
    compte = np.zeros(len(X))
    for _ in range(n):
        compte[np.random.choice(X)] += 1
    
    freqX = compte / n
    plt.bar(X, freqX, width=0.2, color="red")
    plt.show()
    
    [print(str(i) + ": " + str(n * 100) + "%") for i, n in enumerate(freqX, X[0])]
```

```python
plot_observation(1000, [i for i in range(0, 10)])
```

```python
plot_observation(1000, [i for i in range(-1, 3)])
```
